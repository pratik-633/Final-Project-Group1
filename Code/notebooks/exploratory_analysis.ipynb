```python
import pandas as pd
import numpy as np
import torch
import matplotlib.pyplot as plt
import os
import json
```

```python
with open('../../logs/wgan_gp_64_history.json', 'r') as f:
    wgan_64_history = json.load(f)
    
with open('../../logs/wgan_gp_128_history.json', 'r') as f:
    wgan_128_history = json.load(f)
```

```python
epochs = wgan_64_history['train_history']['epoch']
gen_loss = wgan_64_history['train_history']['generator_loss']

plt.figure(figsize=(10, 5))
plt.plot(epochs, gen_loss)
plt.xlabel('Epoch')
plt.ylabel('Generator Loss')
plt.title('WGAN-GP 64 Res Generator Loss Over Training')
plt.show()
```

```python
epochs_128 = wgan_128_history['train_history']['epoch']
gen_loss_128 = wgan_128_history['train_history']['generator_loss']
min_len = min(len(epochs_128), len(gen_loss_128))

plt.figure(figsize=(10, 5))
plt.plot(epochs_128[:min_len], gen_loss_128[:min_len])
plt.xlabel('Epoch')
plt.ylabel('Generator Loss')
plt.title('WGAN-GP 128 Res Generator Loss Over Training')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(wgan_64_history['train_history']['epoch'], wgan_64_history['train_history']['critic_loss'])
plt.xlabel('Epoch')
plt.ylabel('Critic Loss')
plt.title('WGAN-GP 64 Res Critic Loss Over Training')
plt.show()
```

```python
epochs = wgan_128_history['train_history']['epoch']
critic_loss = wgan_128_history['train_history']['critic_loss']
min_len = min(len(epochs), len(critic_loss))

plt.figure(figsize=(10, 5))
plt.plot(epochs[:min_len], critic_loss[:min_len])
plt.xlabel('Epoch')
plt.ylabel('Critic Loss')
plt.title('WGAN-GP 128 Res Critic Loss Over Training')
plt.show()
```

```python
epochs = wgan_64_history['train_history']['epoch']
fid_all = wgan_64_history['train_history']['fid']

fid_values = []
epoch_fid_pairs = []
for i in range(len(epochs)):
    if fid_all[i] is not None:
        fid_values.append(float(fid_all[i]))
        epoch_fid_pairs.append((epochs[i], float(fid_all[i])))

plt.figure(figsize=(10, 5))
plt.plot([pair[0] for pair in epoch_fid_pairs], [pair[1] for pair in epoch_fid_pairs])
plt.xlabel('Epoch')
plt.ylabel('FID Score')
plt.title('WGAN-GP 64 Res FID Score Over Training')
plt.show()
```

```python
epochs = wgan_128_history['train_history']['epoch']
fid_all = wgan_128_history['train_history']['fid']

fid_values = []
epoch_fid_pairs = []
for i in range(len(fid_all)):
    if fid_all[i] is not None:
        fid_values.append(float(fid_all[i]))
        epoch_fid_pairs.append((epochs[i], float(fid_all[i])))

plt.figure(figsize=(10, 5))
plt.plot([pair[0] for pair in epoch_fid_pairs], [pair[1] for pair in epoch_fid_pairs])
plt.xlabel('Epoch')
plt.ylabel('FID Score')
plt.title('WGAN-GP 128 Res FID Score Over Training')
plt.show()
```
