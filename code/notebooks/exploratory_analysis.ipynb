```python
import pandas as pd
import numpy as np
import torch
import matplotlib.pyplot as plt
import os
import json
```

```python
with open('../../logs/wgan_gp_64_history.json', 'r') as f:
    wgan_history = json.load(f)
```

```python
epochs = wgan_history['train_history']['epoch']
gen_loss = wgan_history['train_history']['generator_loss']

plt.figure(figsize=(10, 5))
plt.plot(epochs, gen_loss)
plt.xlabel('Epoch')
plt.ylabel('Generator Loss')
plt.title('WGAN-GP Generator Loss Over Training')
plt.show()
```

```python
epochs = wgan_history['train_history']['epoch']
critic_loss = wgan_history['train_history']['critic_loss']

plt.figure(figsize=(10, 5))
plt.plot(epochs, critic_loss)
plt.xlabel('Epoch')
plt.ylabel('Critic Loss')
plt.title('WGAN-GP Critic Loss Over Training')
plt.show()
```

```python
epochs = wgan_history['train_history']['epoch']
fid_all = wgan_history['train_history']['fid']

fid_values = []
epoch_fid_pairs = []
for i in range(len(epochs)):
    if fid_all[i] is not None:
        fid_values.append(float(fid_all[i]))
        epoch_fid_pairs.append((epochs[i], float(fid_all[i])))

plt.figure(figsize=(10, 5))
plt.plot([pair[0] for pair in epoch_fid_pairs], [pair[1] for pair in epoch_fid_pairs])
plt.xlabel('Epoch')
plt.ylabel('FID Score')
plt.title('WGAN-GP FID Score Over Training')
plt.show()
```
